# DATA VISUALIZATION

### Hist plot (xem phân bố ảnh)

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

# Histogram of height
axes[0].hist(df_covid['height'], bins=30, alpha=0.7, color='b')
axes[0].set_xlabel("Height")
axes[0].set_ylabel("Count")
axes[0].set_title("Distribution of Image Height on SARS-COV-2 dataset ")

# Histogram of width
axes[1].hist(df_covid['width'], bins=30, alpha=0.7, color='g')
axes[1].set_xlabel("Width")
axes[1].set_ylabel("Count")
axes[1].set_title("Distribution of Image Width on SARS-COV-2 dataset")

plt.tight_layout()  
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

# Histogram of height
axes[0].hist(df_brain['height'], bins=30, alpha=0.7, color='b')
axes[0].set_xlabel("Height")
axes[0].set_ylabel("Count")
axes[0].set_title("Distribution of Image Height on Brain tumor dataset ")

# Histogram of width
axes[1].hist(df_brain['width'], bins=30, alpha=0.7, color='g')
axes[1].set_xlabel("Width")
axes[1].set_ylabel("Count")
axes[1].set_title("Distribution of Image Width Brain tumor dataset")

plt.tight_layout()  
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

# Histogram of height
axes[0].hist(df_lung['height'], bins=30, alpha=0.7, color='b')
axes[0].set_xlabel("Height")
axes[0].set_ylabel("Count")
axes[0].set_title("Distribution of Image Height on Lung cancer dataset ")

# Histogram of width
axes[1].hist(df_lung['width'], bins=30, alpha=0.7, color='g')
axes[1].set_xlabel("Width")
axes[1].set_ylabel("Count")
axes[1].set_title("Distribution of Image Width on Lung cancer dataset")

plt.tight_layout()  
plt.show()
```

### Distribution of classes

```python
# Đếm số lượng ảnh theo từng class
class_counts_covid = df_covid['class'].value_counts()
class_counts_brain = df_brain['class'].value_counts()
class_counts_lung = df_lung['class'].value_counts()
# Vẽ biểu đồ tròn

# Tạo figure và các subplot
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Danh sách datasets và tiêu đề tương ứng
datasets = [df_covid, df_brain, df_lung]
titles = ["SARS-COV-2 Dataset","Brain Dataset", "Lung Dataset"]
colors = [['blue', 'green', 'red', 'purple', 'orange'],  # Màu cho từng dataset
          ['cyan', 'magenta', 'yellow', 'gray', 'brown'],
          ['pink', 'lime', 'gold', 'teal', 'navy']]

# Vẽ biểu đồ tròn cho từng dataset
for i, (df, title, color) in enumerate(zip(datasets, titles, colors)):
    class_counts = df['class'].value_counts()
    axes[i].pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', colors=color)
    axes[i].set_title(title)

# Hiển thị toàn bộ biểu đồ
plt.tight_layout()
plt.show()
```

### Scatter 

```python
# Chuyển class sang mã số nếu cần trực quan hóa với scatter plot
plt.scatter(df_brain['width'], df_brain['height'], c=df_brain['class'].astype('category').cat.codes, alpha=0.5)
plt.xlabel("Chiều rộng")
plt.ylabel("Chiều cao")
plt.title("Scatter plot: Chiều rộng vs Chiều cao")
plt.show()
```
