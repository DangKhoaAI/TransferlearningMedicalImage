```python
# Đếm số lượng ảnh theo từng class
class_counts_covid = df_covid['class'].value_counts()
class_counts_brain = df_brain['class'].value_counts()
class_counts_lung = df_lung['class'].value_counts()
# Vẽ biểu đồ tròn

# Tạo figure và các subplot
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Danh sách datasets và tiêu đề tương ứng
datasets = [df_covid, df_brain, df_lung]
titles = ["COVID Dataset", "Brain Dataset", "Lung Dataset"]
colors = [['blue', 'green', 'red', 'purple', 'orange'],  # Màu cho từng dataset
          ['cyan', 'magenta', 'yellow', 'gray', 'brown'],
          ['pink', 'lime', 'gold', 'teal', 'navy']]

# Vẽ biểu đồ tròn cho từng dataset
for i, (df, title, color) in enumerate(zip(datasets, titles, colors)):
    class_counts = df['class'].value_counts()
    axes[i].pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', colors=color)
    axes[i].set_title(title)

# Hiển thị toàn bộ biểu đồ
plt.tight_layout()
plt.show()
```

### Outlier

```python
def detect_outliers(df, column):
    """Hàm phát hiện outlier dựa trên IQR"""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Tính outlier cho từng dataset và từng cột
datasets = {'COVID': df_covid, 'Brain': df_brain, 'Lung': df_lung}
for name, df in datasets.items():
    outliers_height = detect_outliers(df, 'height')
    outliers_width = detect_outliers(df, 'width')
    print(f"Dataset {name}:")
    print(f"- Số outlier (height): {len(outliers_height)}")
    print(f"- Số outlier (width): {len(outliers_width)}\n")
```
